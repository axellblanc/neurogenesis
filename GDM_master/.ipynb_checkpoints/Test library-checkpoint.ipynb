```python
from gammagwr import GammaGWR
import gtls
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds_iris = gtls.IrisDataset(file='iris.csv', normalize=True)
print("%s from %s loaded." % (ds_iris.name, ds_iris.file), '\n \n') 
g = GammaGWR()
g.init_network(ds_iris, 1)
```

```python
num_context = 1 # number of context descriptors
epochs = 1 # epochs per sample for incremental learning
a_threshold = [0.95, 0.9]
beta = 0.7
learning_rates = [0.2, 0.001]
context = True

g.train_ggwr(ds_iris, epochs, a_threshold[0], beta, learning_rates)
```

```python
g.test_gammagwr(ds_iris, test_accuracy=True)
print('test accuracy: ',g.test_accuracy)
gtls.plot_network(g, True, True) 
```
